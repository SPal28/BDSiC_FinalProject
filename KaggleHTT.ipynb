# Import Data


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#make sure to add the .txt file in
df = pd.read_csv('Huntington_Disease_Dataset.csv', header=None)
df.rename(columns=lambda x: df[x][0], inplace=True)
df = df.iloc[1:]
df = df.drop(['Brain_Volume_Loss', 'Functional_Capacity','Chorea_Score'], axis=1)
df.head()
```

```python
numeric_columns = ["Age", "HTT_CAG_Repeat_Length", "HTT_Gene_Expression_Level", "Protein_Aggregation_Level"]
for column in numeric_columns:
  df[column] = pd.to_numeric(df[column], errors='coerce')
```

```python
from sklearn.preprocessing import OneHotEncoder

def one_hot_encoding(df, cat_col):
  df_pandas_encoded = pd.get_dummies(df, columns=cat_col, drop_first=True)
  encoder = OneHotEncoder(sparse_output=False)
  one_hot_encoded = encoder.fit_transform(df[cat_col])
  one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(cat_col))
  df_sklearn_encoded = pd.concat([df.drop(cat_col, axis=1), one_hot_df], axis = 1)
  return df_sklearn_encoded
```

```python
df = one_hot_encoding(df, ['Cognitive_Decline'])
```
